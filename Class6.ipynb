```python
import cv2 as cv
import matplotlib.pyplot as plt
import random
import numpy as np

def fplt(_src):
    plt.imshow(_src)
    plt.show

img = cv.imread("pepega.png")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
fplt(img)
```

```python
# [ Noise ]
ar_op = np.zeros(img.shape, np.uint8)
p = 0.05 # probablity of noise
for i in range (img.shape[0]):
    for j in range(img.shape[1]):
        r = random.random()
        if r < p/2:
            ar_op[i][j] = 0,0,0
        elif r < p:
            ar_op[i][j] = 255,255,255
        else:
            ar_op[i][j] = img[i][j]
fplt(ar_op)
```

```python
# [ A knl for convolution ]
knl = np.ones((8, 8), np.float32)/64
knl
```

```python
ar_op = cv.filter2D(img, -1, knl)
fplt(ar_op)
```

```python
ar_op = cv.boxFilter(img, -1, (24,24), normalize=True)
fplt(ar_op)
```

```python
ar_op = cv.boxFilter(img, -1, (2,2), normalize=False)
fplt(ar_op)
```

```python
ar_op = cv.blur(img, (64,64))
fplt(ar_op)
```
